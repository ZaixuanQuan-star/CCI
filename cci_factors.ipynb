## 划分训练集和测试集

```python
from sklearn.model_selection import train_test_split
# 划分训练集和测试集，测试集占比为0.2（20%）
data_train, data_test = train_test_split(data, test_size=0.2, random_state=42)
# 打印训练集和测试集的形状
print("训练集形状:", data_train.shape)
print("测试集形状:", data_test.shape)
```

```python
X_train=data_train[feas]
y_train=data_train['cci']
X_test=data_test[feas]
y_test=data_test['cci']
```

## 相关系数法

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression

# 使用皮尔逊相关系数选择k个最好的特征
selector = SelectKBest(score_func=f_regression, k=8)
selector.fit(X_train, y_train)

# 获取特征选择后的数据
X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

# 输出选择的特征
selected_indices = selector.get_support(indices=True)
selected_features = X_train.columns[selected_indices]
selected_features
```

## RFE法

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

# Initialize the RFE model with LinearRegression as the estimator
selector = RFE(estimator=LinearRegression(), n_features_to_select=8)

# Fit the model and transform the data
X_train_rfe = selector.fit_transform(X_train, y_train)

# Output the selected features
selected_indices = selector.get_support(indices=True)
selected_features_rfe = X_train.columns[selected_indices]
selected_features_rfe
```

```python
# Initialize the Lasso model with L1 penalty
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)

# Fit the model and transform the data to select features
from sklearn.feature_selection import SelectFromModel
selector = SelectFromModel(lasso)
X_train_selected = selector.fit_transform(X_train, y_train)

# Get the indices of the selected features
selected_indices = selector.get_support(indices=True)

# Output the selected features
selected_features_emb = X_train.columns[selected_indices]
print(selected_features_emb)
```

```python
fea1=set(selected_features)
fea2=set(selected_features_rfe)
fea3=set(selected_features_emb)
```

```python
fea_all=fea1.union(fea2).union(fea3)
fea_all
```

## 线性回归模型


```python
#训练模型
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train)
#预测
y_pred = model.predict(X_test_scaled)
#评估模型
from sklearn.metrics import mean_squared_error, r2_score
mse_linear = mean_squared_error(y_test, y_pred)
r2_linear = r2_score(y_test, y_pred)
print("Mean Squared Error:", mse_linear)
print("R2 Score:", r2_linear)
#输出各变量的系数
coef_linear = model.coef_
print("Coefficients:", coef_linear)
print(len(coef_linear))
```

## 决策树

```python
from sklearn.tree import DecisionTreeRegressor

# Create a new instance of the DecisionTreeRegressor class
model = DecisionTreeRegressor(max_depth=4, min_samples_split=8)

# Fit the model to the training data
model.fit(X_train_scaled, y_train)

# Use the model to make predictions on the test data
y_pred = model.predict(X_test_scaled)

# Calculate the mean squared error of the predictions
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate the R-squared score of the model
r2 = r2_score(y_test, y_pred)
print("R-squared Score:", r2)

important_features_decisiontree=model.feature_importances_
print('important_features:',important_features_decisiontree)
```

```python
import torch
import torch.nn as nn

class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to('mps')
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to('mps')

        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])
        return out
    
# 设置超参数
input_size = X_train_scaled.shape[1]
hidden_size = 64
num_layers = 2
output_size = 1
num_epochs = 100
learning_rate = 0.001

# 初始化模型、损失函数和优化器
model = LSTMModel(input_size, hidden_size, num_layers, output_size).to('mps')
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# 转换数据为张量并移到MPS设备上
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32).unsqueeze(1).to('mps')
y_train_tensor = torch.tensor(y_train.to_numpy(), dtype=torch.float32).unsqueeze(1).to('mps')
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32).unsqueeze(1).to('mps')

# 训练模型
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 10 == 0:
        print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')

# 在测试集上进行预测
model.eval()
with torch.no_grad():
    y_pred_tensor = model(X_test_tensor)

# 将预测结果转换回原始数据尺度
y_pred = y_pred_tensor.cpu().numpy().reshape(-1, 1)
y_test_original = y_test.to_numpy().reshape(-1, 1)
import numpy as np
# 计算RMSE
rmse = np.sqrt(mean_squared_error(y_test_original, y_pred))
print(f'Test RMSE: {rmse}')
#计算r2
r2 = r2_score(y_test_original, y_pred)
print(f'Test R2: {r2}')
```
